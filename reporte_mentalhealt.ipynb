### Paso 3: Limpieza de Datos

```python
# porcentaje de datos que faltan en cada variable
porcentaje_perdidos = (df_mental_health.isnull().sum() / len(df_mental_health)) * 100
print(porcentaje_perdidos)
```

```python
# porcentaje de datos que faltan en todo el Dataset
valores_perdidos_total = df_mental_health.isnull().sum()
registros_total = np.prod(df_mental_health.shape)
total_nulos = valores_perdidos_total.sum()
porcentaje_perdido = (total_nulos/registros_total)*100
print(porcentaje_perdido)
```

```python
# Eliminamos las columnas que tienen un número importante de valores nulos
df_mental_health_clean = df_mental_health.drop(['Medication_Use','Substance_Use'], axis=1)
df_mental_health_clean.head()
```

### Paso 4: Análisis Estadístico

```python
# Estadísticas de Anxiety_Score
Anxiety_Score_mean = df_mental_health_clean['Anxiety_Score'].mean()
Anxiety_Score_median = df_mental_health_clean['Anxiety_Score'].median()
Anxiety_Score_std = df_mental_health_clean['Anxiety_Score'].std()

# Estadísticas de Depression_Score
Depression_Score_mean = df_mental_health_clean['Depression_Score'].mean()
Depression_Score_median = df_mental_health_clean['Depression_Score'].median()
Depression_Score_std = df_mental_health_clean['Depression_Score'].std()

# Estadísticas de Stress_Level
Stress_Level_mean = df_mental_health_clean['Stress_Level'].mean()
Stress_Level_median = df_mental_health_clean['Stress_Level'].median()
Stress_Level_std = df_mental_health_clean['Stress_Level'].std()

print("Estadísticas de Anxiety_Score:")
print("Media:", Anxiety_Score_mean)
print("Mediana:", Anxiety_Score_median)
print("Desviación estándar:", Anxiety_Score_std)

print("\nEstadísticas de Depression_Score:")
print("Media:", Depression_Score_mean)
print("Mediana:", Depression_Score_median)
print("Desviación estándar:", Depression_Score_std)

print("\nEstadísticas de Stress_Level:")
print("Media:", Stress_Level_mean)
print("Mediana:", Stress_Level_median)
print("Desviación estándar:", Stress_Level_std)
```

### Paso 5: Filtrado de Datos

```python
# La variable Anxiety_Score no presenta valores mayores a 20 por lo que se hace la equivalencia a 100, obteniendo 16 como un equivalente a 80

df_Anxiety_Score_alto = df_mental_health_clean[df_mental_health_clean['Anxiety_Score'] > 16]
df_Anxiety_Score_alto.head()
```

```python
# Creamos una nueva columna en donde clasificamos con base en las horas de sueño
df_Anxiety_Score_alto.loc[df_Anxiety_Score_alto['Sleep_Hours'] < 5, 'Calidad_suenio'] = 'Malo'
df_Anxiety_Score_alto.loc[(df_Anxiety_Score_alto['Sleep_Hours'] >= 5) & (df_Anxiety_Score_alto['Sleep_Hours'] <= 7), 'Calidad_suenio'] = 'Regular'
df_Anxiety_Score_alto.loc[df_Anxiety_Score_alto['Sleep_Hours'] > 7, 'Calidad_suenio'] = 'Bueno'
df_Anxiety_Score_alto.head()
```

```python
# Porcentaje total que presenta una ansiedad alta
porcentaje_anxiety_score_alto = (len(df_Anxiety_Score_alto) / len(df_mental_health_clean)) * 100
print(porcentaje_anxiety_score_alto)
```

```python
# Revisamoos si existe correlación entre la variable Anxiety_Score y la variable Sleep_Hours
correlacion_Anxiety_Sleep_Hours = df_mental_health_clean['Anxiety_Score'].corr(df_mental_health_clean['Sleep_Hours'])
print(correlacion_Anxiety_Sleep_Hours)
```
